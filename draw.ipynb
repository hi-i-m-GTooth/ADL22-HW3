```python
import json
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
rlt_dir = "/tmp2/yzliu/git/ADL22-HW3/preds/"
```

```python
x = []
r1 = []
r2 = []
rl = []
ds = sorted(os.listdir(rlt_dir), key=lambda x: int(x.replace("ep", "")))
for d in ds:
    full = os.path.join(rlt_dir, d, "score.json")
    d = d.replace("ep", "")
    x.append(int(d)+1)

    score = json.load(open(full))
    r1.append(score["rouge-1"]['f'])
    r2.append(score["rouge-2"]['f'])
    rl.append(score["rouge-l"]['f'])
```

```python
x
```

```python
plt.plot(x, r1, label="rouge-1")
plt.plot(x, r2, label="rouge-2")
plt.plot(x, rl, label="rouge-l")
plt.legend(loc="best")
plt.title("Learning curves")
plt.xlabel("epoch")
plt.ylabel("F1 score")
plt.show()
```
